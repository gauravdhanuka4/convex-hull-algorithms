```python
import random as random
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.animation as FuncAnimation
```

```python
stack = []
randomlist_x = []
for i in range (0,70):
    n = random.randint(0,100)
    randomlist_x.append(n)

x = randomlist_x

randomlist_y = []
for i in range (0,70):
    n = random.randint(0,100)
    randomlist_y.append(n)
    
y = randomlist_y
```

```python
points = list(zip(x,y))
print(points)
points.sort(key=lambda x:[x[0],x[1]])
print()
print(points)
print()
```

```python
stack = []
start = points.pop(0)
print(points)
print(stack)
stack.append(start)
print(stack)
print()
print()
```

```python
def slope(p1, p2):
    if p1[0] == p2[0]:
        return float('inf')
    else:
        return 1.0*(p1[1]-p2[1])/(p1[0]-p2[0])
```

```python
points.sort(key=lambda p: (slope(p,start),-p[0],p[1]))
print(points)
```

```python
def get_cross_product(p1,p2,p3):
    return ((p2[0] - p1[0])*(p3[1] - p1[1])) - ((p2[1] - p1[1])*(p3[0] - p1[0]))
```

```python
points.sort(key=lambda p: (slope(p,start), -p[1],p[0]))
i = 0
def convex_hull():
    for pt in points:
        stack.append(pt)
        print(stack) 
        while len(stack) > 2 and get_cross_product(stack[-3],stack[-2],stack[-1]) < 0:
            stack.pop(-2)
            #print(stack)
            
convex_hull()
print()
print()
print()
#print(stack_1)
```

```python
x_hull,y_hull = np.array(stack).T.tolist() 
print(x_hull)
print(y_hull)
x_hull.append(x_hull[0])
y_hull.append(y_hull[0])
```

```python
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))
ax[1].plot(x_hull,y_hull)
ax[1].scatter(x,y,alpha = 1)
ax[1].set_title('Convex Hull of 70 random points in 100x100 grid')

ax[0].set_title('Random 70 points in 100x100 grid')
ax[0].scatter(x,y,alpha = 1)
fig.show()
```
